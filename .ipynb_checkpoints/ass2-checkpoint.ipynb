```python
# lets explore movies.csv
m_cols = ['movie_id','movie title','release date','video release date','IMDb URL'
          ,'unknown','Action','Adventure','Animation','Childrens','Comedy',
          'Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi',
        'Thriller','War','Western']
data= pd.read_csv('C:/Users/moran/Desktop/ml-100k/ml-100k/u.item',delimiter='|',names=m_cols,encoding='latin-1')
data.head()
```

```python
#explore users
u_cols = ['user_id','age','gender','occupation','zip_code']
data_u = pd.read_csv('C:/Users/moran/Desktop/ml-100k/ml-100k/u.user',delimiter='|',names=u_cols,encoding='latin-1')
data_u.head()
```

```python
#explore ratings
r_cols = ['user_id','movie_id','rating','timestamp']
data_r = pd.read_csv('C:/Users/moran/Desktop/ml-100k/ml-100k/u1.base',delimiter='\t',names=r_cols,encoding='latin-1')
data_r.head()
```

```python
#merging two dataframes movies and ratings - Q2 section a
# calculate averege rating and sort in descending order 
# display the 3 movies with the highest rating
movie_data_ratings_data=data.merge(data_r,on = 'movie_id',how = 'inner')
movie_data_ratings_data_1 = movie_data_ratings_data.groupby(['movie_id','movie title'])['rating'].mean().sort_values(ascending=False).reset_index()
movie_data_ratings_data_1.columns = ['movie_id','movie title', 'avg rating']
movie_data_ratings_data_1.head(3)
```

```python
#Q2 section b
# calculate count rating and sort in descending order
# display the 3 movies with the highest rating
movie_data_ratings_data_2 = movie_data_ratings_data.groupby(['movie_id','movie title'])['rating'].size().sort_values(ascending=False).reset_index()
movie_data_ratings_data_2.columns = ['movie_id','movie title', 'num rating']
movie_data_ratings_data_2.head(3)
```

```python
#displays high rated movies - Q2 section c
# calculate % rating and sort in descending order 
# display the 3 movies with the highest rating
high_rated= movie_data_ratings_data['rating']>4.0
count_ratings_more_then_4_per_movie = movie_data_ratings_data.loc[high_rated].groupby('movie_id').size().sort_values(ascending=False).reset_index()
count_ratings_more_then_4_per_movie.columns = ['movie_id','count rating > 4']
percentage_data = pd.merge(count_ratings_more_then_4_per_movie, movie_data_ratings_data_2,on=['movie_id'])
percentage_data['%  rating'] = (percentage_data['count rating > 4'] / percentage_data['num rating']) *100
percentage_data.columns = percentage_data.columns.to_series().apply(lambda x: x.strip())
percentage_data=percentage_data.sort_values(by='%  rating',ascending=False)
percentage_data.head(3)
```

```python
#Q2 section d - out measure: Weighted Rating
#Weighted Rating (WR) = ((v/(v+m))*R)+((m/(v+m))*C)
#v is the number of votes for the movie
#m is the minimum votes required to be listed in the chart
#R is the average rating of the movie
#C is the mean vote across the whole report

movie_data_ratings_3 = movie_data_ratings_data_1.merge(movie_data_ratings_data_2,on = ['movie_id','movie title'],how = 'left')
movie_data_ratings_3 = movie_data_ratings_3[['movie_id','movie title', 'num rating', 'avg rating']]
R = movie_data_ratings_3['avg rating']
v = movie_data_ratings_3['num rating']
m = (movie_data_ratings_data['rating'].count())/2
C = movie_data_ratings_data['rating'].mean()
movie_data_ratings_3['Weighted Rating'] = ((v/(v+m))*R)+((m/(v+m))*C)
movie_data_ratings_3.columns = movie_data_ratings_3.columns.to_series().apply(lambda x: x.strip())
movie_data_ratings_3=movie_data_ratings_3.sort_values(by='Weighted Rating',ascending=False)
movie_data_ratings_3.head(3)
```

```python
#Q2 section e
# merge two dataframe users and movies with rating
# distribution of information to men and women
user_merged_with_movie_ratings=movie_data_ratings_data.merge(data_u,on = 'user_id',how = 'inner')
female_aggregetion = user_merged_with_movie_ratings['gender'] == 'F'
male_aggregetion = user_merged_with_movie_ratings['gender'] == 'M'
data_of_female = user_merged_with_movie_ratings.loc[female_aggregetion]
data_of_male = user_merged_with_movie_ratings.loc[male_aggregetion]
```

```python
#Q3 - section a
# explore test set and train set 
# calculate the average of all rating for mae
# calculate avetage rating foe each movie
# merge test set and train set
train_cols = ['user_id','movie_id','rating','timestamp']
train_set = pd.read_csv('C:/Users/moran/Desktop/ml-100k/ml-100k/u1.base',delimiter='\t',names=train_cols,encoding='latin-1')
test_cols = ['user_id','movie_id','rating','timestamp']
test_set= pd.read_csv('C:/Users/moran/Desktop/ml-100k/ml-100k/u1.test',delimiter='\t',names=test_cols,encoding='latin-1')
avg_rating_of_all_train = train_set['rating'].mean()
train_set = train_set.groupby(['movie_id'])['rating'].mean().reset_index()
train_set.columns = ['movie_id','avg rating']
data_merged = train_set.merge(test_set,on ='movie_id')
data_merged.head()
```

```python
#build a prediction model based on the average rating of each movie
#calculate mae
MAE_data_merged = 0
for index, row in data_merged.iterrows():
    pi = row['avg rating']
    ri = row['rating']
    MAE_data_merged = MAE_data_merged + abs(pi-ri)
N = len(data_merged)   
MAE_data_merged = MAE_data_merged/N
MAE_data_merged
```

```python
# build a prediction model based onaverage rating general all over the movies
# calculate mae
MAE_test_set = 0
for index, row in test_set.iterrows():
    ri = row['rating']
    MAE_test_set = MAE_test_set + abs(avg_rating_of_all_train-ri)
N = len(test_set)   
MAE_test_set = MAE_test_set/N
MAE_test_set
```

The second method has a higher MAE

```python
#calculate precision : Ns / (Ns + Nis) for 2.a - average rating
precision_table_avg = pd.DataFrame(columns=['user_id','precision'])
for user_id in users:
    user_aggregetion = recomended_movies_for_user['user_id'] == user_id
    recomended_movies = recomended_movies_for_user.loc[user_aggregetion]
    recomended_movies = recomended_movies['movie_id'].unique().tolist()
    user_aggregetion_test_set = test_set['user_id'] == user_id
    movies_he_saw = test_set.loc[user_aggregetion_test_set]
    movies_he_saw = movies_he_saw['movie_id'].unique().tolist()
    intersection_movies = list(set(recomended_movies).intersection(movies_he_saw))
    precision = len(intersection_movies) / float(20)
    precision_table_avg = precision_table_avg.append({'user_id':user_id,'precision': precision},ignore_index=True)
precision_table_avg.head()
```
